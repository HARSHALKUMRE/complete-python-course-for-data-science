## Nested Functions

```python
def f():
    def g():
        print("Inside function g")
    g()
    print("Inside function f")
```

```python
f()
```

```python
g()
```

```python
def f():
    def g():
        print("Inside function g")
        f()
    
    g()
    print("Inside function f")
```

```python
f()
```
